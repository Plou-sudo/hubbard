# Implementación del método de Kohn-Sham para el dímero de Hubbard

```python
%matplotlib inline
import numpy as np
from qutip import *
import matplotlib.pyplot as plt
from sympy.utilities.iterables import multiset_permutations
from time import time
import random
```

```python
def ind_menor(vector):
    
    ind_min=[]
    ordenados=sorted(vector)
    
    for i in range (len(vector)):
        for k in range(len(vector)):
        
            if (vector[k]==ordenados[i]):
                
                ind_min.append(k)
                ind_min.append(k)
            
    return ind_min
```

```python
P=2
N_e=2

T=np.zeros([P,P])
V=np.zeros([P,P])
W=np.zeros([P,P])
v=np.zeros(P)

for i in range (0,P-1):
    T[i,i+1]=T[i+1,i]=-0.5

v_ext=np.linspace(0,10,50)

u=[0.2, 1., 2., 5., 10.]   
```

```python
t_0=time()

alpha=0.9

E=[]
delta_N=[]

for i in range (0, len(u)):
    
    en=[]
    delta_n=[]
    
    for j in range (len(v_ext)):
               
        for k in range (0,P):
            v[k]=(-1)**k*v_ext[j]
        
        for k in range (0,P):
                V[k,k]=v[k]
        
        N_2=0       
        epsilon=0   
        
        n_out=np.zeros(P)
        n_old=[random.random() for x in range(P)]
        n_new=[random.random() for x in range(P)]
        conver=0
        
        while (all(abs(np.array(n_old)-np.array(n_out))>0.0000001)):
        #for l in range (300):
        #while(abs(n_new[0]+n_new[1]-n_old[0]-n_old[1])>0.0000001):
       
            for k in range (0, P):
                n_old[k]=n_new[k]
                
            #print(n_old, n_new)
            
            for k in range (0,P):
                W[k,k]=0.5*u[i]*n_old[k]
                
            H_ks=T+V+W
            
            evals, ekets=np.linalg.eig(H_ks)
            
            n_out=np.zeros(P)
            
            for k in range(0,P):
                for e in range(0, N_e):
                    n_out[k]+=ekets[ind_menor((evals))[e]][k]**2
            
            for k in range(0,P):
                n_new[k]=alpha*n_old[k]+(1-alpha)*n_out[k]
            
        
        for k in range(0, P):
            N_2+=n_new[k]**2
            
        for e in range(0, N_e):
            epsilon+=evals[ind_menor(evals)[e]]
                                
        en.append(epsilon-0.25*u[i]*N_2)
        delta_n.append(abs(n_new[0]-n_new[1]))
        
    E.append(en)
    delta_N.append(delta_n)
    
t_1=time()-t_0
    
data = np.column_stack([E[0], E[1], E[2], E[3], E[4]])
np.savetxt('Energia_DFT.txt' , data, fmt=['%.8f','%.8f','%.8f','%.8f','%.8f']) 
```
